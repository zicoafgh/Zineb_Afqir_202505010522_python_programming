## Activity 1: Identify the Components


### 1.1 Inputs
- quiz_1 (first quiz mark entered by user)
- quiz_2 (second quiz mark entered by user)
- quiz_3 (third quiz mark entered by user)
- choice (user input to continue or stop)

### 1.2 Process
- Calculate the average of the three quiz marks
- Check if the average is greater than or equal to 50
- Repeat the process if the user wants to enter another student

### 1.3 Output
- Display the average mark
- Display 'Pass' if average >= 50
- Display 'Fail' if average < 50
- Display 'Program Ended' when user exits

## Activity 2: Design the Algorithm


### 2.1 Pseudocode
```
START
  SET choice = 'y'
  WHILE choice == 'y':
    INPUT quiz_1
    INPUT quiz_2
    INPUT quiz_3
    CALCULATE average = (quiz_1 + quiz_2 + quiz_3) / 3
    DISPLAY average
    IF average >= 50:
      DISPLAY 'Pass'
    ELSE:
      DISPLAY 'Fail'
    INPUT choice 'Continue? Y/N'
END
```

### 2.2 Algorithm Structure
- **Sequence:** The program runs step by step by inputing quiz marks → calculate average → display result
- **Selection:** `if average >= 50` → displays Pass, `else` → displays Fail
- **Iteration:** `while choice == 'y'` → repeats the whole process for another student
